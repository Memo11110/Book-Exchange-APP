```python
import pickle
import torch
from torch.utils.data import Dataset

# Dummy class to match the one saved in the pickle
class StreamspotDataset(Dataset):
    def __init__(self, host_features, net_features, labels):
        self.host_features = host_features
        self.net_features = net_features
        self.labels = labels
        self.num_hosts = len(host_features)
        self.seq_len = host_features[0].shape[1]
        self.host_dim = host_features[0].shape[2]
        self.net_dim = net_features.shape[1]

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return [h[idx] for h in self.host_features], self.net_features[idx], self.labels[idx]

# Redirect pickle loader to our local class
class RedirectUnpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == "utils.loaddata" and name == "StreamspotDataset":
            return StreamspotDataset
        return super().find_class(module, name)

# Load the pickle file
pkl_path = "/content/APT-detect/data/streamspot/sbustreamspot-data-master/snapshots.pkl"
with open(pkl_path, "rb") as f:
    data = RedirectUnpickler(f).load()

print("✅ Loaded:", type(data))
if isinstance(data, dict):
    print("Top-level keys:", list(data.keys())[:5])
elif isinstance(data, list):
    print("List length:", len(data))
    print("Sample types:", [type(i) for i in data[:5]])
```

```python
# ✅ Imports
import pickle
import torch
from torch.utils.data import Dataset
from dgl.dataloading import GraphDataLoader
import dgl
import networkx as nx
import matplotlib.pyplot as plt

# ✅ Define dummy class to allow unpickling
class StreamspotDataset(Dataset):
    def __init__(self, *args, **kwargs):
        pass  # Will be populated by pickle

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.graphs[idx], self.labels[idx]

# ✅ Redirect for pickle loading
class RedirectUnpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == "utils.loaddata" and name == "StreamspotDataset":
            return StreamspotDataset
        return super().find_class(module, name)

# ✅ Load the pickled dataset
pkl_path = "/content/APT-detect/data/streamspot/sbustreamspot-data-master/snapshots.pkl"
with open(pkl_path, "rb") as f:
    raw_dataset = RedirectUnpickler(f).load()

# ✅ Extract and rebuild dataset
attrs = raw_dataset.__dict__
clean_dataset = StreamspotDataset.__new__(StreamspotDataset)
clean_dataset.graphs = attrs["graphs"]
clean_dataset.labels = attrs["labels"]

print("✅ Dataset loaded with", len(clean_dataset), "graphs")

# ✅ Use DGL GraphDataLoader
loader = GraphDataLoader(clean_dataset, batch_size=4, shuffle=True)
batched_graph, labels = next(iter(loader))
print("🔹 Batched graph:", type(batched_graph))
print("🔹 Batch size:", batched_graph.batch_size)
print("🔹 Labels:", labels.tolist())

# ✅ Unbatch and visualize the first 4 graphs
graphs = dgl.unbatch(batched_graph)

fig, axs = plt.subplots(2, 2, figsize=(12, 12))
axs = axs.flatten()

for i, g in enumerate(graphs[:4]):
    nx_g = dgl.to_networkx(g)
    axs[i].set_title(f"Graph {i} - {g.num_nodes()} nodes")
    nx.draw(nx_g, ax=axs[i], node_size=30, with_labels=False)

plt.tight_layout()
plt.show()
```

```python
import pickle
import torch
from torch.utils.data import Dataset
from dgl.dataloading import GraphDataLoader
import dgl
import networkx as nx
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# ✅ Define dummy class to allow unpickling
class StreamspotDataset(Dataset):
    def __init__(self, *args, **kwargs):
        pass  # Will be populated by pickle

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.graphs[idx], self.labels[idx]

# ✅ Redirect for pickle loading
class RedirectUnpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == "utils.loaddata" and name == "StreamspotDataset":
            return StreamspotDataset
        return super().find_class(module, name)

# ✅ Load the pickled dataset
pkl_path = "/content/APT-detect/data/streamspot/sbustreamspot-data-master/snapshots.pkl"
with open(pkl_path, "rb") as f:
    raw_dataset = RedirectUnpickler(f).load()

# ✅ Extract and rebuild clean dataset object
attrs = raw_dataset.__dict__
clean_dataset = StreamspotDataset.__new__(StreamspotDataset)
clean_dataset.graphs = attrs["graphs"]
clean_dataset.labels = attrs["labels"]

print("✅ Dataset loaded with", len(clean_dataset), "graphs")

# ✅ Truncate to a very small subset
MAX_SAMPLES = 8  # use only 8 samples
clean_dataset.graphs = clean_dataset.graphs[:MAX_SAMPLES]
clean_dataset.labels = clean_dataset.labels[:MAX_SAMPLES]
print("✅ Dataset truncated to", len(clean_dataset), "graphs")

# ✅ Use DGL GraphDataLoader
loader = GraphDataLoader(clean_dataset, batch_size=4, shuffle=True)
batched_graph, labels = next(iter(loader))
print("🔹 Batched graph:", type(batched_graph))
print("🔹 Batch size:", batched_graph.batch_size)
print("🔹 Labels:", labels.tolist())

# ✅ Unbatch and visualize the first 4 graphs
graphs = dgl.unbatch(batched_graph)

fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs = axs.flatten()

for i, g in enumerate(graphs[:4]):
    nx_g = dgl.to_networkx(g)
    axs[i].set_title(f"Graph {i} — {g.num_nodes()} nodes")
    nx.draw(nx_g, ax=axs[i], node_size=30, with_labels=False)

plt.tight_layout()
plt.show()
```
